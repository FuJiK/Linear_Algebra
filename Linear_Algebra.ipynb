```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
left = [0,6]
equation1 = [5/3,-1/3]
equation2 = [-1/3,11/3]
plt.plot(left,equation1,color='blue',label='x+3y=5')
plt.plot(left,equation2,color='red',label='2x-3y=1')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```

```python
left = [0,6]
equation1 = [5/3,-1/3]
equation2 = [5/3,-1/3]
plt.plot(left,equation1,color='blue',label='x+3y=5')
plt.plot(left,equation2,color='red',linestyle='dotted',linewidth=3,label='2x+6=10')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.show()
```
