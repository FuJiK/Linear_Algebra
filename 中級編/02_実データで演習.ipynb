# 訓練データ（train）と検証データ（test）を分割

- 実務に出る際には、まだ問題がある。  
- 分けることも、機械学習の一部

例)駄目なケース
- 受験勉強用に10年分の過去問を買ってきた。
- 10年分で勉強（学習）
- 10年分で実力テスト（検証）

例）いいケース

- 前半の5年分で勉強（学習）←訓練データ
- 後半の5年分で実力テスト（検証）←検証データ


```python
# sklearn 内にmode_selectionというライブラリがあって、その中にtrain_test_splitというモジュールがある。
from sklearn.model_selection import train_test_split
```

```python
#　訓練データと検証データの分割
# Shift + Tabで関数の説明を見ることが出来る。
# train_test_split(
# ...     X, y, test_size=0.33　テストの割合（33%を検証用データ）
# , random_state=42) ランダムに（前半5年分、後半5年分とするのではなく、
# ランダムに5年分を訓練データとして用い、残った5年分のデータを検証用データとする）
# 時間軸に関係ないものだと、全てから満遍なくデータを取れたほうがいいので、ランダムにデータを取る。
# ランダムに取る場合は、毎回結果が変わってしまうためランダムの根っこの部分（seed）を固定する必要がある。

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.4,random_state=1)# random_stateは乱数のシードを固定（再現性の確保）
# 一度作ればコピペ
```

```python
# モデルの学習（パラメータの調整）
model.fit(X_train,y_train)
```

```python
# 検証 ← 検証データに対して、どの程度うまく言ってるか。
# Trainで用いていないデータである、Testデータを用いてscoreを出す。
model.score(X_test,y_test)
```

- こちらが、八百長してない実力テストの実際の結果となる。


- 0.74とか0.72でいいのか？ということは実際のオペレーション
- オペレーションに入れる際にどこまで精度を求めているのか次第

```python
# 検証 ← 訓練データ
model.score(X_train,y_train)
# 基本的には、訓練データの方が精度がよく、それに対し検証データは訓練のデータに対してパラメータの調整を行っているので
# 当然、訓練データはベストな結果が出てくる！
# ただ、0.74から0.72など数値が近ければ学習はうまくいってる
```

# パラメータの確認

- その変数が重要なのか、どうなのか？
- ビジネスを行う上で、うまく予測できたけどなぜそれがうまく予測できたのか？ということを知りたい時がある。
- そういったケースのときに、なぜうまくいってますか？というのを聞かれて機械学習でなんとなく、、、、だとよろしくない。
- そういったときのヒントがこの重みにある。
- 重みが大きければ大きいほど、変数に影響を与えている！
- 重みの大きさを見てあげれば、一つなぜうまくいってますかという回答に「この変数が効いているようです」と答えることが出来、ヒントになることがある。

```python
# 重みのところ
# 調整後のパラメータwの値 
# x1~x13に対してのパラメータwの値
# 係数
model.coef_
```

```python
# Tips 有効桁数の設定
np.set_printoptions(precision=3)
```

```python
# 有効桁数の設定　小数点以下3桁に設定
model.coef_
```

```python
# 指数関数での表記の禁止
np.set_printoptions(precision=3,suppress=True)
```

```python
# 有効桁数の設定　小数点以下3桁に設定
# 指数関数での表記の禁止
# パット見でどれが大きくてどれが小さいかがわかる。
model.coef_
```

# 値の見方

- 0.067　パラメータの値がめっちゃ小さいな。（0に近いな。）
- -17.205 絶対値で見ると、めっちゃ大きいな重要な値なのか！と結論づけたりする。

```python
# (重み×入力変数 )+ (重み×入力変数 )+...+(重み×入力変数 )
df.head(3)
# x1の値が0.00632に対して0.09を掛けてる
# x2の値が18に対して、0.067を掛けてる
# 入力変数の値によるスケールが結構違ったりする
# かけ合わせたものに対するスケールの大きさが重要！であって、”model.coef_”で重みを単体で見ても変数の値も単体で違うので、
# 議論に持ってくることが出来ない。
```

重み（パラメータ$w$のこと）を見るだけでは、どの変数が影響を与えているのかがわからない。

- 解決策としては、確率統計！

ノイズ的な要素でいらないような変数、いらないようなデータが入ってしまっているときにどのようにして取り除くのか？
というところも含めて、解決策を知った上で重回帰分析の問題に戻ってくる！
